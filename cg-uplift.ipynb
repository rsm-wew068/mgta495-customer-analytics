## Q2

```python
evars = cg_rct_stacked.columns.drop(['converted', 'ad', 'rnd_30k', 'training'])
evars
```

```python
clf_treatment = rsm.model.logistic(
    data={'stacked': cg_rct_stacked.query('training == 1 & ad == 1')},
    rvar='converted',
    lev='yes',
    evar=evars,
)
clf_treatment.summary()
```

```python
logit_treat_t5 = clf_treatment.plot('vimp', ret=True).index[0:5]
```

```python
clf_treatment.plot('pred', incl=logit_treat_t5)
```

Under the treatment case, we see that the number of ads clicked is the most important model. But this was expected under the prior case we saw. Overall, we see most of the variables have a smooth, positive effect on the response variable.

```python
clf_control = rsm.model.logistic(
    data={'stacked': cg_rct_stacked.query('training == 1 & ad == 0')},
    rvar='converted',
    lev='yes',
    evar=evars,
)
clf_control.summary()
```

```python
logit_control_t5 = clf_control.plot('vimp', ret=True).index[0:5]
```

```python
tab_up_logit = rsm.uplift_tab(cg_rct_stacked.query('training == 0'), 'converted', 'yes', 'uplift_score', 'ad', 1, qnt=20)
tab_up_logit
```

As we target more customers, the incremental uplift steadily increases until around **80-85%** of the population, where it reaches its peak at approximately **8% uplift**. After this point, the incremental uplift starts to decline, suggesting that targeting additional customers beyond this threshold leads to diminishing returns, indicating that ***the campaign is most effective when focusing on the top 80-85% of ranked customers***.

```python
fig = rsm.uplift_plot(cg_rct_stacked.query('training == 0'), 'converted', 'yes', 'uplift_score', 'ad', 1, qnt=20)
```

The highest-ranked segments show a strong positive uplift, with the top bins reaching up to **30% uplift**, meaning the ads significantly improve conversion rates for these customers. As we move to lower-ranked segments, the uplift gradually declines, showing reduced ad effectiveness. Toward the lowest-ranked bins, the uplift becomes negative, with values dropping as low as **-40%**, suggesting that ads may have a counterproductive effect on these users, possibly due to ad fatigue or targeting users who were unlikely to purchase in the first place.

Overall, the results suggest that targeting approximately **80-85%** of the highest-propensity customers is the optimal strategy. Including more than this threshold would lead to lower efficiency and wasted ad spend, while targeting fewer would mean missing out on potential conversions.

# Neural Net

## Q2

### Control

```python
from itertools import product
import numpy as np

hls = []
for size in range(1,3):
    hls.extend(product(range(1,6), repeat=size))

param_grid = {
    'hidden_layer_sizes': hls,
    'alpha': np.arange(0, 1.1, 0.1),
}
scoring = {'AUC': 'roc_auc'}
```

```python
nn_control = rsm.model.mlp(
    data={'stacked': cg_rct_stacked.query('training == 1 & ad == 0')},
    rvar='converted',
    lev='yes',
    evar=evars,
    hidden_layer_sizes=(1,)
)
nn_control.summary()
```

```python
nn_control_t5 = nn_control.plot("vimp_sklearn", ret=True).index[0:5]
```

```python
nn_control.plot('pred', incl=nn_control_t5)
```

```python
from sklearn.model_selection import StratifiedKFold, GridSearchCV
```

```python
cv_nn_control_file = 'cv-objects/nn_control.pkl'
if os.path.exists(cv_nn_control_file):
    cv_nn_control = rsm.load_state(cv_nn_control_file)['cv_nn_control']
else:
    strat_k_fold = StratifiedKFold(n_splits=5, shuffle=True, random_state=1234)
    cv_nn_control = GridSearchCV(
        nn_control.fitted,
        param_grid,
        scoring=scoring,
        cv=strat_k_fold,
        n_jobs=-1,
        refit=list(scoring.keys())[0],
        verbose=5
    ).fit(nn_control.data_onehot, nn_control.data.converted)
    if not os.path.exists('cv-objects'):
        os.mkdir('cv-objects')
    rsm.save_state({'cv_nn_control': cv_nn_control}, cv_nn_control_file)
```

```python
cv_nn_control.best_params_, cv_nn_control.best_score_
```

```python
pd.DataFrame(cv_nn_control.cv_results_).sort_values('rank_test_AUC').head()
```

```python
nncv_control = rsm.model.mlp(
    data={'stacked': cg_rct_stacked.query('training == 1 & ad == 0')},
    rvar='converted',
    lev='yes',
    evar=evars,
    **cv_nn_control.best_params_
)
nncv_control.summary()
```

```python
nncv_control_t5 = nncv_control.plot("vimp", ret=True).index[0:5]
```

```python
nncv_control.plot('pred', incl=nncv_control_t5)
```

Initially, **NumSpaceHeroBadges** was the strongest predictor, indicating that gameplay achievements played a major role in conversion. However, after hyperparameter tuning, **NumFriends** became the most significant factor, suggesting that social engagement is a more critical driver of conversions. Additionally, the importance of **NumAdsClicked** decreased, showing that ad engagement alone is not enough to predict conversions effectively.

With these refinements, the model’s AUC improved from **0.841** to **0.886**, demonstrating better predictive power. This means we now have a more accurate model for uplift prediction, allowing us to target customers based on deeper behavioral insights rather than relying too heavily on direct ad engagement or gameplay achievements.

For our main two reasons we suspect the following:
* Under the case that a propensity model out performs a uplift, the including the "for sure" buyers in the targeting scheme actually provides a strong source of profit for the firm. Whereas a greater positive results for the uplift model means that only focusing marketing efforts on the right population saves marketing costs, which leads to bigger margins and effective marketing campaigns.
* Alternatively, we have to consider that having a relatively small breakeven point (0.1) provides a larger margin of error in propensity models. Under an uplift case, we want to target everyone who will more than a 10 percent bump in receiving an ad effect. This is key because we are considering the effects of marketing rather than flat out probability of person buying. Propensity does not consider negative conditions of giving an ad, in other words, when the conditions flip to say $P(\text{buy}|\text{no ad})$ is greater than $P(\text{buy}|\text{ad})$.
